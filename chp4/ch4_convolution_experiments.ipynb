```python
import torch
from torch.nn import Conv1d
```

```python
batch_size = 2
one_hot_size = 10
sequence_width = 7
data = torch.randn(batch_size, one_hot_size, sequence_width)
```

```python
data.shape
```

```python
data
```

```python
conv1 = Conv1d(in_channels=one_hot_size, out_channels=16,
               kernel_size=3, bias=False)
intermediate1 = conv1(data)
```

```python
intermediate1.shape
```

```python
intermediate1
```

```python
channels00 = data[0,:,:3]
channels00
```

```python
k00 = conv1.weight[0]
k00
```

```python
channels00 * k00
```

```python
torch.sum(channels00 * k00)
```
